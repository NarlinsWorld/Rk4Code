Imagine you want to look at the value of some output y(t)

obtained from a differential system
$$\frac{dy(t)}{dt}+\frac{y(t)}{4}=x(t)\tag{3}$$

You first need to rearrange this to be an equation for $dy(t)/dt$
as a function of t, y, and constants. This equation also has an independent $x(t)$

so that will be a part of the function as well:
$$\frac{dy(t)}{dt}=x(t)−\frac{y(t)}{4} \tag{4}$$
Now if you further assume that:
$$x(t)=cos(3t) \tag{5}$$ $$y(t)=−1 \tag{6}$$

you can numerically solve this differential equation (and plot the results for both x(t)
and y(t)) with: 

```python
# %% Imports
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from numpy import linspace,sin, cos

# %% Define independent function and derivative function
def x(t):
    return cos(3 * t)

def f(t, y, c):
    dydt = x(t) - y / 4
    return dydt

# %% Define time spans, initial values, and constants
tspan = linspace(0, 15, 1000)
yinit = [-1]
c = []

# %% Solve differential equation
sol = solve_ivp(lambda t, y: f(t, y, c),
                [tspan[0], tspan[-1]], yinit, t_eval=tspan)
# %% Plot independent and dependent variable
# Note that sol.y[0] is needed to extract a 1-D array
plt.figure(1)
plt.clf()
fig, ax = plt.subplots(num=1)
ax.plot(sol.t, x(sol.t), 'k-', label='Input')
ax.plot(sol.t, sol.y[0], 'k--', label='Output')
ax.legend(loc='best')
```
