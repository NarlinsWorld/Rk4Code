```python
from scipy import linspace, sin
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def pos(x):
    return max(x, 0)

def neg(x):
    return max(-x, 0)

a, b, L, M = 17, 1, 15.4, 0.75

def system(t, z):
    y, yp = z # yp = y'
    return [yp, 10 + L*sin(M*t) - 0.01*yp - a*pos(y) + b*neg(y)]

t = linspace(0, 100, 301)

sol = solve_ivp(system, [0, 100], [1, 0], t_eval=t)
plt.plot(sol.t, sol.y[0])
plt.xlabel("$t$")
plt.ylabel("$y$")
```
